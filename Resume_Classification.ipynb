```python
import numpy as np
import tensorflow as tf
import gradio as gr

from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
texts = [
    "python machine learning data analysis deep learning",
    "java spring boot backend development microservices",
    "recruitment hiring onboarding hr policies",
    "sales marketing business development client handling",
    "html css javascript frontend react ui design",
    "data science python pandas numpy visualization",
    "employee engagement hr operations talent acquisition",
    "android development kotlin mobile app",
    "deep learning neural networks ai research",
    "marketing seo content strategy branding"
]

labels = [
    "Data Science",
    "Software Developer",
    "HR",
    "Sales",
    "Web Developer",
    "Data Science",
    "HR",
    "Software Developer",
    "Data Science",
    "Sales"
]
```

```python
label_dict = {label: i for i, label in enumerate(set(labels))}
reverse_label_dict = {i: label for label, i in label_dict.items()}

y = np.array([label_dict[label] for label in labels])
```

```python
tokenizer = Tokenizer(num_words=1000)
tokenizer.fit_on_texts(texts)

X = tokenizer.texts_to_sequences(texts)
X = pad_sequences(X, maxlen=10)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model = Sequential([
    Embedding(input_dim=1000, output_dim=64, input_length=10),
    LSTM(64),
    Dense(32, activation='relu'),
    Dense(len(label_dict), activation='softmax')
])

model.compile(loss='sparse_categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

model.summary()
```

```python
model.fit(X_train, y_train, epochs=20, verbose=1)
```

```python
def predict_resume(text):
    seq = tokenizer.texts_to_sequences([text])
    padded = pad_sequences(seq, maxlen=10)

    pred = model.predict(padded)
    label = reverse_label_dict[np.argmax(pred)]

    return f"Predicted Category: {label}"
```

```python
interface = gr.Interface(
    fn=predict_resume,
    inputs=gr.Textbox(lines=5, placeholder="Paste resume text here..."),
    outputs="text",
    title="Resume Classification App",
    description="Classifies resume into job category using Deep Learning",
    theme=gr.themes.Soft()
)

interface.launch()
```
